```python
import pandas as pd
df=pd.read_csv('Illusive Invoices Data Requirement 2018-2020 for Gal.csv')
```

```python
df.head()
```

```python
df['Type']=df['Type'].apply(lambda x: x.strip())
df['Type']=df['Type'].str.lower()
df['Type'].value_counts()
```

```python
df.loc[df['Type']=='6 year','Type']='6 years'
df['Type'].value_counts()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Type'].value_counts().to_frame().index.to_list()
```

```python
month_labal={'other':1,'monthly':1,'quarterly':3,'annual':12,'2 years':24,'3 years':36,'4 years':48,'5 years':60,'6 years':72}
```

```python
df_new=pd.DataFrame(columns=df.columns.to_list())
df_new
```

```python
from datetime import date, timedelta
import calendar
for index, row in df.iterrows():
    df_new=df_new.append(row).reset_index(drop=True)
    no_copy=month_labal[row.Type]
    for i in range(1,no_copy):
        row_number=len(df_new)
        start_date = date(df_new.loc[row_number-1,'Date'].year, df_new.loc[row_number-1,'Date'].month, df_new.loc[row_number-1,'Date'].day)
        days_in_month = calendar.monthrange(start_date.year, start_date.month)[1]
        end=start_date + timedelta(days=days_in_month)
        df_new=df_new.append(row).reset_index(drop=True)
        df_new.loc[row_number,'Date']=end
        
        
```
